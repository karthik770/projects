```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline


headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv('Desktop//car.txt',names=headers)
df.head(2)
```

```python
df.replace('?',np.nan,inplace=True)
df
#df.isnull().sum()
```

```python
print(df.shape)
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df

df['normalized-losses'].replace(np.nan,df['normalized-losses'].astype('float').mean()
,inplace=True)
df['stroke'].replace(np.nan,df['stroke'].astype('float').mean(),inplace=True)
df['bore'].replace(np.nan,df['bore'].astype('float').mean(),inplace=True)
df['peak-rpm'].replace(np.nan,df['horsepower'].astype('float').mean(),inplace=True)
df['horsepower'].replace(np.nan,df['peak-rpm'].astype('float').mean(),inplace=True)
df1=df
df1
```

```python
df1.isnull().sum()
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df["normalized-losses"] = df["normalized-losses"].astype("int")
df["price"] = df["price"].astype("float")
df["peak-rpm"] = df["peak-rpm"].astype("float")
df['horsepower']=df['horsepower'].astype('float')
```

```python
print(df['fuel-type'].value_counts())

fuel_variable=pd.get_dummies(df['fuel-type'])
fuel_variable.head(2)
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
df1.isna().sum()
```

```python
df = pd.concat([df,fuel_variable],axis=1)
df
```

```python
df = df.drop('fuel-type', axis=1)
df.head(2)
```
