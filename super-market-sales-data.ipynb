```python
n=int(input(''))
m=int(input(''))
print('Entered n:',n ,'\n', 'Entered m',m)

matrix=[]
for i in range(n):
    row=list(map(int,input('').split()))
    matrix.append(row)
   
    
    
transpose=[]
for j in range(m):
    col=[]
    for i in range(n):
        col.append(matrix[i][j])
    transpose.append(col)
        
print('\n'.join([' '.join(map(str, row)) for row in transpose]))

            
            
flatten = []
for i in range(n):
    for j in range(m):
        flatten.append(matrix[i][j])
        
        
print(' '.join(map(str,flatten)))
    

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "C:\\Users\\Raju\\Desktop\\data set\\supermarket_sales - Sheet1.csv"
sales_data = pd.read_csv(url)
sales_data.head(2)
```

```python
sales_data.columns = sales_data.columns.str.replace(' ', '_')
sales_data.columns
```

```python
sales_data['Customer_type']
```

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'], format='%d,%m,%Y')
sales_data['Date']


```

```python
print(sales_data.isnull().sum())
```

```python
sales_data.drop_duplicates(inplace=True)
```

```python
a=sales_data['Branch'].value_counts()
print(a)
sns.countplot(x='Branch', data=sales_data)
plt.title('Number of Sales by Branch')
```

```python
print(sales_data['Customer_type'].value_counts())
sns.countplot(x='Customer_type',data=sales_data)
plt.title('Number of Sales by Customer Type')
```

```python
print(sales_data['Gender'].value_counts())
sns.countplot(x='Gender', data=sales_data)
plt.title('Number of Sales by Gender')
```

```python
print(sales_data['Product_line'].value_counts())
sns.countplot(x='Product_line', data=sales_data)
plt.xticks(rotation=80)
plt.title('Number of Sales by Product Line')
```

```python
plt.figure(figsize=(25,13))
sns.countplot(sales_data['Rating'] )
plt.xticks(rotation=90)
plt.title('Distribution of Unit Prices')
```

```python
print(sales_data['Unit_price'].value_counts())
plt.figure(figsize=(20,12))
plt.hist(sales_data['Unit_price'],bins=30)
plt.show()
```

```python
plt.hist(sales_data['Quantity'], bins=10)
plt.title('Distribution of Quantity')
```

```python
d = sales_data['City'].value_counts()
print(d)
label = sales_data['City'].value_counts().index

plt.pie(d , labels=label, autopct='%1.1f%%')
plt.title('Distribution of Sales by City')
```
